```python
# Function to convert pixel values from a string format to an array format.
def string2array(x):
  return np.array(x.split(' ')).reshape(48,48,1).astype('float32')
```

```python
# Resize images from (48, 48) to (96, 96)
def resize(x):
  img = x.reshape(48,48)
  # Using OpenCV's resize function to resize the img (which is now 48x48) to a new size of 96x96 pixels.
  # The interpolation=cv2.INTER_CUBIC argument tells OpenCV to use cubic interpolation, a higher-quality method for resizing that works well when enlarging images.
  return cv2.resize(img, dsize=(96,96), interpolation=cv2.INTER_CUBIC)
```

```python
facial_expression_df[' pixels'] = facial_expression_df[' pixels'].apply(lambda x: string2array(x))
```

```python
facial_expression_df[' pixels'] = facial_expression_df[' pixels'].apply(lambda x: resize(x))
```
